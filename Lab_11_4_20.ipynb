```python
import numpy as np

def rule1(c, erA):
    ret = abs(c) * erA
    return ret

def rule2(c, m, A, erA):
    ret = abs(c*m*A**(m-1)) * erA
    return ret

def rule3(erA, erB):
    ret = np.sqrt(erA**2 + erB**2)
    return ret
    
def rule4(m, n, A, B, erA, erB):
    ret = np.sqrt((m*(erA/A))**2 + (n * (erB/B)**2))
    return ret

def boyount(p, V):
    ret = (p) * 9.8 * V
    return ret

def volume(r, h):
    ret = 3.1415926 * (r**2) * h
    return ret
```

```python
#Excercise 1
heights = [ .01915, .0382, .0569, .0772]
radius = .012775 #m 
volumes = [0, 0, 0, 0]
forces = [0, 0, 0, 0]
erVolumeCyl = [0, 0, 0, 0]

force1 = boyount(1000, volume(.01, .05/4))
force2 = boyount(1000, volume(.01, .05/2))
force3 = boyount(1000, volume(.01, (.05 * .75)))
force4 = boyount(1000, volume(.01, .05))
print(volume(.01, .05/4))
print(volume(.01, .05/2))
print(volume(.01, .05*.75))
print(volume(.01, .05))
print('Excercise boyount force1: ', force1)
print('Excercise boyount force2: ', force2)
print('Excercise boyount force3: ', force3)
print('Excercise boyount force4: ', force4)

#Part 1
#1
massWat = .4834 - .1617
denWat = massWat / 0.000326 #m^3
erDens = rule4(1, 1, massWat, .000326, .0001, 0.000001) * denWat
print('\n')
print('Density of water: ',denWat)
print('Density error: ', erDens)
print('\n')
#2
for i in range(4):
    volumes[i] = volume(radius, heights[i])
    print("Volumes", volumes[i])
    
print('\n')

for i in range(4):
    forces[i] = boyount(denWat, volumes[i])
    print("Forces: ", forces[i])

print("\n")

for i in range(4):
    erVolumeCyl[i] = rule4(2, 1, radius, heights[i], .00005, .00005) * volume(radius, heights[i])
    print("Err in volumes Cyl:", erVolumeCyl[i])

print("\n")
for i in range(4):
    uncertainty = rule4(1, 1, denWat, volumes[i], erDens, erVolumeCyl[i]) * forces[i]
    print('Uncertainty', i, ":", uncertainty)
```
